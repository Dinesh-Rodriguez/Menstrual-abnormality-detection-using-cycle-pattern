```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

def detect_anomalies(user_df, threshold=5):
    user_df = user_df.sort_values('cycle_start_date')
    series = user_df['cycle_length_days'].astype(float)
    
    try:
        # Fit ARIMA model (simple AR(1))
        model = ARIMA(series, order=(1,0,0))
        model_fit = model.fit()
        
        # Forecast cycle lengths
        forecast = model_fit.predict(start=0, end=len(series)-1)
        
        # Residuals = difference between actual and forecast
        residuals = series - forecast
        
        # Anomalies = residuals exceeding threshold
        anomalies = user_df[(residuals > threshold) | (residuals < -threshold)]
        
        # Plot
        plt.figure(figsize=(10,5))
        plt.plot(user_df['cycle_start_date'], series, label='Actual')
        plt.plot(user_df['cycle_start_date'], forecast, label='Forecast')
        plt.scatter(anomalies['cycle_start_date'], anomalies['cycle_length_days'], color='red', label='Anomaly', zorder=5)
        plt.title(f'User {user_df.user_id.iloc[0]} Cycle Lengths')
        plt.xlabel('Cycle Start Date')
        plt.ylabel('Cycle Length (days)')
        plt.legend()
        plt.show()
        
        return anomalies
    except Exception as e:
        print(f"ARIMA failed for user {user_df.user_id.iloc[0]}: {e}")
        return pd.DataFrame()
```

```python

from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```python
# Sort by user and date
time_series_df['cycle_start_date'] = pd.to_datetime(time_series_df['cycle_start_date'])
time_series_df = time_series_df.sort_values(['user_id', 'cycle_start_date'])

# Select features for LSTM
features = ['cycle_length_days', 'pain_score', 'bleeding_volume_score', 'pattern_disruption_score']

# Scale features to [0,1]
scalers = {}
for f in features:
    scalers[f] = MinMaxScaler()
    time_series_df[f] = scalers[f].fit_transform(time_series_df[[f]])
```

```python
seq_length = 3  # use last 3 cycles to predict next
def create_sequences(df, seq_length=3):
    X, y = [], []
    for i in range(len(df)-seq_length):
        X.append(df[features].iloc[i:i+seq_length].values)
        y.append(df['cycle_length_days'].iloc[i+seq_length])
    return np.array(X), np.array(y)

# Store sequences per user
user_sequences = {}
for uid in time_series_df['user_id'].unique():
    user_df = time_series_df[time_series_df['user_id']==uid]
    X, y = create_sequences(user_df, seq_length)
    if len(X) > 0:
        user_sequences[uid] = (X, y)
```

```python
X_all, y_all = [], []
for X, y in user_sequences.values():
    X_all.append(X)
    y_all.append(y)
X_all = np.vstack(X_all)
y_all = np.hstack(y_all)
```

```python
model = Sequential([
    LSTM(32, input_shape=(seq_length, len(features)), return_sequences=False),
    Dense(16, activation='relu'),
    Dense(1)  # predict next cycle length
])

model.compile(optimizer=Adam(0.001), loss='mse')
model.summary()
```

```python
history = model.fit(X_all, y_all, epochs=50, batch_size=16, validation_split=0.1)
```

```python
threshold = 0.1  # normalized difference threshold

anomalies = []
for uid, (X, y_true) in user_sequences.items():
    y_pred = model.predict(X).flatten()
    residuals = np.abs(y_pred - y_true)
    for i, r in enumerate(residuals):
        if r > threshold:
            anomalies.append({
                'user_id': uid,
                'cycle_number': i + seq_length + 1,
                'predicted_norm': y_pred[i],
                'actual_norm': y_true[i],
                'residual_norm': r
            })

anomalies_df = pd.DataFrame(anomalies)
```

```python
anomalies_df['predicted_days'] = scalers['cycle_length_days'].inverse_transform(anomalies_df[['predicted_norm']])
anomalies_df['actual_days'] = scalers['cycle_length_days'].inverse_transform(anomalies_df[['actual_norm']])
anomalies_df['residual_days'] = anomalies_df['actual_days'] - anomalies_df['predicted_days']

print(anomalies_df[['user_id','cycle_number','predicted_days','actual_days','residual_days']])
```

```python
from sklearn.metrics import mean_squared_error
y_pred = model.predict(X_all).flatten()
print("MSE:", mean_squared_error(y_all, y_pred))
```
